### OUTLIERS

```python
#outliers
# Select numerical columns
numerical_cols = final_merged.select_dtypes(include=['int64', 'float64']).columns

# Define the number of rows and columns dynamically
num_cols = len(numerical_cols)
num_rows = (num_cols // 3) + (num_cols % 3 > 0) 

# Create boxplots
plt.figure(figsize=(15, num_rows * 4))  
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(num_rows, 3, i)  
    sns.boxplot(data=final_merged, x=col)
    plt.title(f'Boxplot of {col} (After Capping)')
plt.tight_layout()
plt.show()
```

#### Detection and Correction of Outliers

```python
#detect and correct 
contVar = ['temp', 'atemp', 'weathersit', 'windspeed', 'casual', 'registered', 'cnt']

for i in contVar:
    Q1 = np.quantile(final_merged[i], 0.25)  
    Q3 = np.quantile(final_merged[i], 0.75) 
    IQR = Q3 - Q1  
    LB = Q1 - 1.5 * IQR  
    UB = Q3 + 1.5 * IQR  
    
    # Cap values at lower and upper bounds
    final_merged[i] = np.where(final_merged[i] < LB, LB, final_merged[i])
    final_merged[i] = np.where(final_merged[i] > UB, UB, final_merged[i])
```

#### After correction of outliers

```python
#again check 
# Select numerical columns
numerical_cols = final_merged.select_dtypes(include=['int64', 'float64']).columns

# Define the number of rows and columns dynamically
num_cols = len(numerical_cols)
num_rows = (num_cols // 3) + (num_cols % 3 > 0)  

# Create boxplots
plt.figure(figsize=(15, num_rows * 4))  
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(num_rows, 3, i)  
    sns.boxplot(data=final_merged, x=col)
    plt.title(f'Boxplot of {col} (After Capping)')
plt.tight_layout()
plt.show()
```

### Skewness detection

```python
# Select numerical columns
numerical_cols = final_merged.select_dtypes(include=['int64', 'float64']).columns

# Compute skewness
skewness_values = final_merged[numerical_cols].skew()

# Display skewness values
print("Skewness of numerical features:\n", skewness_values)

#>1 or <-1	Highly skewed data (needs transformation)
#<-0.5 (Negative Skewed)	Tail is on the left (left-skewed)
#>0.5 (Positive Skewed)	Tail is on the right (right-skewed)
```

#### Skewness visualisation

```python
skewed_cols = ['weathersit', 'casual', 'registered', 'cnt',]

plt.figure(figsize=(10, 6))
for i, col in enumerate(skewed_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(final_merged[col], kde=True, bins=30)
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

#### Correction of skewness

```python

for col in ['weathersit', 'casual']:
    final_merged[col] = np.log1p(final_merged[col])  
```

```python
for col in ['registered', 'cnt']:
    final_merged[col] = np.sqrt(final_merged[col])
```
