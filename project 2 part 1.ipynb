### Step.8: **Merging dataset1 and dataset2**

```python
# Merge dataset1 and dataset2 on 'instant' using an inner join
merged_df = pd.merge(df, df2, on="instant", how="inner")
```

### Step.7: Exploratory Data Analysis

```python
#descriptive statistics
merged_df.describe()
```

#### histogram

```python
# Define relevant numerical columns
num_cols = ["temp", "atemp", "hum", "windspeed", "casual", "registered", "cnt"]

# Plot histograms for selected numerical columns
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=merged_df, x=col,bins=20,kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

#### scatterplot

```python
# Define relevant numerical pairs for scatter plots
scatter_cols = [
    ("temp", "cnt"),
    ("atemp", "cnt"),
    ("hum", "cnt"),
    ("windspeed", "cnt"),
    ("registered", "cnt"),
    ("casual", "cnt"),
]

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 8))  
axes = axes.flatten()  

# Loop through columns and plot on subplots
for index, (x_col, y_col) in enumerate(scatter_cols):
    sns.scatterplot(data=merged_df, x=x_col, y=y_col, alpha=0.5, ax=axes[index])
    axes[index].set_title(f"{x_col} vs {y_col}")
    axes[index].set_xlabel(x_col)
    axes[index].set_ylabel(y_col)

plt.tight_layout() 
plt.show()
```
